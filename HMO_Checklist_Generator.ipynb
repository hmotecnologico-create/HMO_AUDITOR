# 📋 HMO Auditor — Generador de Checklist de Requerimientos

Este Notebook automatiza la creación de la **Guía de Preparación Documental** en formato PDF. 

### 🎯 Objetivo:
Proporcionar al cliente (la empresa auditada) un documento formal que detalle qué archivos debe preparar, por qué son necesarios (base normativa) y cómo deben ser creados.

### 🤖 Agentes Involucrados:
- **Agente de Normatividad:** Define la lista de documentos basados en ISO 9001, 14001, 27001.
- **Motor de Composición PDF:** Utiliza la librería `fpdf2` para generar el archivo con identidad corporativa de HMO.

## 🛠️ 1. Importación de Librerías y Módulos
Cargamos el generador de PDF profesional (`HMO_PDF_Generator`) que contiene las plantillas maestras.

```python
import os
import sys

# Asegurar que podemos importar desde las carpetas del proyecto
sys.path.append(os.path.join(os.getcwd(), 'HMO_Auditor_Master_V1', '04_Arquitectura_y_Diseno', 'Scripts_Generadores'))

try:
    from HMO_PDF_Generator import generate_preparation_guide_pdf
    print("✅ Módulo HMO_PDF_Generator cargado correctamente.")
except ImportError as e:
    print(f"❌ Error al cargar el generador: {e}")
```

## 📚 2. Base de Conocimiento de Requerimientos
Aquí definimos los documentos estándar para una auditoría de **Calidad (ISO 9001:2015)**.

```python
REQ_CALIDAD = [
    {
        "doc": "Camara de Comercio",
        "justificacion": "ISO 9001:4.1 - Requisito para entender el contexto legal de la organización.",
        "instrucciones": "Descargue el certificado de la Cámara de Comercio con vigencia no mayor a 30 días."
    },
    {
        "doc": "RUT",
        "justificacion": "Identificación tributaria necesaria para la trazabilidad legal.",
        "instrucciones": "Obtenga el PDF actualizado desde el portal Muisca de la DIAN."
    },
    {
        "doc": "Misión y Visión",
        "justificacion": "ISO 9001:5.2 - Base para establecer la política de calidad y dirección estratégica.",
        "instrucciones": "Redacte el propósito de la empresa y su proyección a futuro en un documento formal."
    },
    {
        "doc": "Organigrama",
        "justificacion": "ISO 9001:5.3 - Define roles, responsabilidades y autoridades en la organización.",
        "instrucciones": "Dibuje la estructura jerárquica de la empresa detallando cada área."
    }
]

print(f"Cargados {len(REQ_CALIDAD)} requerimientos base de Calidad.")
```

## 🚀 3. Generación del PDF
Ejecutamos la función que compila los datos en un documento profesional. 

**Nota:** El archivo se guardará en la carpeta `PDF_Pruebas` de este directorio.

```python
EMPRESA_TEST = "Innovatech Solutions SAS"
OUTPUT_DIR = os.path.join(os.getcwd(), 'PDF_Pruebas')

if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

path_generado = generate_preparation_guide_pdf(
    company_name=EMPRESA_TEST,
    output_path=OUTPUT_DIR,
    doc_requirements=REQ_CALIDAD,
    norma="ISO 9001:2015"
)

print(f"✅ ¡Documento generado con éxito!")
print(f"📍 Ubicación: {path_generado}")
```

## 🔍 4. Explicación del Código Interno

### La clase `HMO_PDF`
Utiliza herencia de la librería `FPDF`. 
- **`header()`**: Se ejecuta al inicio de cada página. Inserta el logo (pendiente) y el título con fuente *Helvetica Bold*.
- **`footer()`**: Se ejecuta al final. Inserta el número de página e información de confidencialidad.

### La función `generate_preparation_guide_pdf`
Coordina la lógica:
1. Crea la instancia del PDF.
2. Itera sobre la lista de requerimientos.
3. Aplica colores estandarizados de HMO (Azul Elite).
4. Usa `multi_cell` para manejar textos largos sin que se salgan de la página.
