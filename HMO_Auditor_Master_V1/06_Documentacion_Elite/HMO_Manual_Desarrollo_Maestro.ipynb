## 🛠️ Fase 1: Arquitectura y Justificación Técnica

**Justificación:** Se seleccionó Ollama para ejecutar modelos como Llama 3 o Mistral en local. Esto permite:

## ⚙️ Fase 2: Desarrollo de Generadores Legales

La validez jurídica de un formato de auditoría depende de su integridad. Implementamos generadores que inyectan metadatos y sellos de integridad SHA-256.

### Código de Ejemplo: Generación con Blindaje Judicial
A continuación se muestra la lógica simplificada de cómo el sistema sella cada documento.

```python
import hashlib

def generate_integrity_hash(content):
    """Genera el sello SHA-256 para el blindaje legal del documento."""
    sha256 = hashlib.sha256()
    sha256.update(content.encode('utf-8'))
    return sha256.hexdigest()

# Ejemplo de simulación de sellado
document_text = "Programa de Auditoria - Empresa Innovatech - 2026"
file_hash = generate_integrity_hash(document_text)
print(f"Sello de Integridad Generado: {file_hash}")
```
